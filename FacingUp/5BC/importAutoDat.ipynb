Bifurcation diagram

```python
bT = pd.read_table("b.mu", skiprows=16, delim_whitespace=True, header=None)
bT = bT.apply(pd.to_numeric, errors='coerce')
bT
```

```python
#bT.loc[bT[0] == 2][5].tolist()
#bInd = 4;
mpl.rcParams.update(mpl.rcParamsDefault)
plt.rc("font", family="Roboto Condensed")
plt.rc("xtick", labelsize="small")
plt.rc("ytick", labelsize="small")

fig = plt.figure(figsize=(8, 2), dpi=100)
ax = plt.subplot(xlim=[-60, 20], ylim=[-np.pi-0.1, np.pi+0.1])

n = 7
for bInd in np.arange(1,n):
    linewidth = 1.5 if bInd == 1 else 1
    linestyle = "-" if bInd == 1 else "-"
    color = "C1" if bInd == 1 else "%.2f" % (bInd / n)

    ax.plot(bT.loc[bT[0] == bInd][4].tolist(),
            bT.loc[bT[0] == bInd][9].tolist(),
            color=color,
            clip_on=False,
            zorder=n - bInd,
            linewidth=linewidth,
            linestyle=linestyle,
            )
plt.show()
```

```python
pdbif = pd.read_table("b.mu", index_col=0, skiprows=16, delim_whitespace=True, header=None)
pdbif['nCnt']=pdbif.isnull().sum(axis=1)
pdbif = pdbif[pdbif['nCnt']==3]
pdbif = pdbif.dropna(axis='columns')
cols = pdbif.columns[pdbif.dtypes.eq('object')]
pdbif[cols] = pdbif[cols].apply(pd.to_numeric, errors='coerce')
pdbif
```
